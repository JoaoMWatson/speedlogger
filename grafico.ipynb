```python
import pandas as pd
import matplotlib
import matplotlib.pylab as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
df = pd.read_csv("log.csv").head(30)
print(df)
```

```python
labels = df['Timestamp']
download = df['Download']
upload = df['Upload']

x = np.arange(len(labels))

width = 0.2

fig, ax = plt.subplots()
rect1 = ax.bar(x - width/2, download, width, label='Download', align='center')
rect2 = ax.bar(x + width/2, upload, width, label='Upload', align='center')

ax.set_ylabel('Mb')
ax.set_title('MORRA CLARO')
ax.set_xticks(x)
ax.set_xticklabels(labels)

plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='center')

ax.legend()

fig.tight_layout()
```
