```python
import pandas as pd
import matplotlib
import matplotlib.pylab as plt
import numpy as np
```

```python
df = pd.read_csv("log.csv")
print(df)
```

```python
labels = df['Timestamp']
download = df['Download']
upload = df['Upload']

x = np.arange(len(labels))

width = 0.2

fig, ax = plt.subplots()
rect1 = ax.bar(x - width/2, download, width, label='Download')
rect2 = ax.bar(x + width/2, upload, width, label='Upload')

ax.set_ylabel('Mb')
ax.set_title('MORRA CLARO')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
